```python
import pandas as pd
```

```python
# ------------ 연도별 어린이교통사고 데이터 정제 ------------
def modify_csv(df, area_df):
    df = df.sort_values(by=['발생건수'], axis=0, ascending=False)
    df = df.reset_index(drop=True)

    area_df = area_df[['지역','발생건수', '사망자수', '부상자수']]
    # SZ_## = 스쿨존 내 데이터
    area_df.columns = ['지역', 'SZ_발생건수', 'SZ_사망자수', 'SZ_부상자수']
    area_df = area_df.sort_values(by=['SZ_발생건수'], axis=0, ascending=False)
    area_df = area_df.reset_index(drop=True)
    
    data = pd.merge(df, area_df, on='지역', how='left')
    # 도로 위험도 지수 계산
    risk_index = []
    area_risk_index = []
    gap = []
    for i in range(len(data)):
        risk_index.append(((data['사망자수'][i]*12+data['부상자수'][i]*3)/data['발생건수'][i]) + 0.001)      
        if data['SZ_발생건수'][i] != 0:
            area_risk_index.append(((data['SZ_사망자수'][i]*12+data['SZ_부상자수'][i]*3)/data['SZ_발생건수'][i]) + 0.001)
        else:
            area_risk_index.append(0)
    data['hazard'] = risk_index
    data['SZ_hazard'] = area_risk_index
    
    for i in range(len(data)):
        gap.append(data['hazard'][i] - data['SZ_hazard'][i])
        
    data['hazard gap'] = gap
    data = data.sort_values(by=['hazard gap'], axis=0, ascending=False)
    data = data.round(2)
    
    return data
```

```python
df = pd.read_excel('child_acc_19.xlsx')
area_df = pd.read_excel('childarea_acc_19.xlsx')
```

```python
# 2019년도 교통사고 데이터 정제
data = modify_csv(df, area_df)
```

```python
data = data[['지역', 'hazard', 'SZ_hazard', 'hazard gap']]
```

```python
data.columns = ['지역', '2019_risk', '2019_C_risk', '2019_gap']
```

```python
data = data.reset_index(drop=True)
```

```python
data.to_csv('2019_지수.csv', index=False)
```
