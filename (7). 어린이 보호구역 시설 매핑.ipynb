```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('서울시 어린이보호구역 위치.csv', encoding='cp949')
```

```python
df = df[['위도', '경도', '대상시설명', '관리기관명']]
```

```python
df.head()
```

```python
for i in range(len(df)):    
    m = folium.Map(
    location=[df['위도'][i],df['경도'][i]],
    zoom_start=13)
    
for i in range(len(df)):
    folium.Marker(location=[df['위도'][i],df['경도'][i]],icon=folium.Icon(color='red',icon='star')).add_to(m)

m.save('map.html')
```
