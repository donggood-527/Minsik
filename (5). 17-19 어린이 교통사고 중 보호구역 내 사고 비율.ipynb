```python
import pandas as pd
```

```python
def ratio(df, area_df):
    data = pd.merge(df, area_df, on='지역', how='left')
    area_acc = []
    for i in range(len(data)):
        area_acc.append((data['발생건수_y'][i]/data['발생건수_x'][i])*100)
    
    data['ratio_2017'] = area_acc
    data_2017 = data.round(2)
    return data_2017
```

```python
df = pd.read_excel('child_acc_19.xlsx')
area_df = pd.read_excel('childarea_acc_19.xlsx')
```

```python
data_2019 = ratio(df, area_df)
```

```python
df = pd.read_excel('child_acc_18.xlsx')
area_df = pd.read_excel('childarea_acc_18.xlsx')
```

```python
data_2018 = ratio(df, area_df)
```

```python
df = pd.read_excel('child_acc_17.xlsx')
area_df = pd.read_excel('childarea_acc_17.xlsx')
```

```python
data_2017 = ratio(df, area_df)
```

```python
data_2017.head()
```

```python
data_2019 = data_2019[['지역', 'ratio_2019']]
data_2018 = data_2018[['지역', 'ratio_2018']]
data_2017 = data_2017[['지역', 'ratio_2017']]
```

```python
data = pd.merge(data_2018, data_2019, on='지역', how='left')
data = pd.merge(data_2017, data, on='지역', how='left')
```

```python
data.head()
```

```python
data.set_index(data['지역'], inplace=True)
```

```python
data = data.drop(['지역'],axis=1)
data = data.drop(['합계'], axis=0)
```
