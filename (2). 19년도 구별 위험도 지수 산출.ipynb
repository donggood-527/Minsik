```python
import pandas as pd
```

```python
# 발생건수 기준 정렬
df = pd.read_excel('child_acc_19.xlsx')
df = df.sort_values(by=['발생건수'], axis=0, ascending=False)
df = df.reset_index(drop=True)

area_df = pd.read_excel('childarea_acc_19.xlsx')
area_df = area_df.sort_values(by=['발생건수'], axis=0, ascending=False)
area_df = area_df.reset_index(drop=True)
```

```python
df.head()
```

```python
area_df.head()
```

```python
area_df = area_df[['지역','발생건수', '사망자수', '부상자수']]
area_df.columns = ['지역', 'area_발생건수', 'area_사망자수', 'area_부상자수']
```

```python
# 어린이 교통사고 데이터, 어린이보호구역 내 교통사고 데이터 병합
df = pd.merge(df, area_df, on='지역', how='left')
```

```python
df.head()
```

```python
# ----------- 위험도지수 산출 -----------
risk_index = []
area_risk_index = []

for i in range(len(df)):
    risk_index.append((df['사망자수'][i]*12+df['부상자수'][i]*3)/df['발생건수'][i])
    area_risk_index.append((df['area_사망자수'][i]*12+df['area_부상자수'][i]*3)/df['area_발생건수'][i])
print(risk_index)
print('-------')
print(area_risk_index)
df['risk_index'] = risk_index
df['area_risk_index'] = area_risk_index
```

```python
gap = []
for i in range(len(df)):
    if df['area_risk_index'][i] != 'nan':
        gap.append(df['risk_index'][i] - df['area_risk_index'][i])
    elif df['area_risk_index'][i] == 0.000000:
        gap.append('-')
df['gap'] = gap
```

```python
df = df.sort_values(by=['gap'], axis=0, ascending=False)
df = df.reset_index(drop=True)
df
```
