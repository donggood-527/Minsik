```python
import pandas as pd
```

```python
df_2019 = pd.read_csv('risk지수.csv')
df_2018 = pd.read_csv('risk지수_2018.csv')
df_2017 = pd.read_csv('risk지수_2017.csv')
```

```python
df_2019 = df_2019[['지역', 'hazard gap']]
df_2018 = df_2018[['지역', 'hazard gap']]
df_2017 = df_2017[['지역', 'hazard gap']]
```

```python
data = pd.merge(df_2019, df_2018, on='지역', how='left')
data = pd.merge(data, df_2017, on='지역', how='left')
data
```

```python
third_hazard = []
for i in range(len(data)):
    third_hazard.append((data['hazard gap_x'][i] + data['hazard gap_y'][i] + data['hazard gap'][i])/3)
print(third_hazard)
```

```python
data['3year_hazard'] = third_hazard
```

```python
data = data[['지역', '3year_hazard']]
data
```

```python
data.to_csv('3년치 risk_index.csv', encoding='utf-8', index=False)
```
