```python
import pandas as pd
```

```python
plt_df = pd.read_csv('구별 시설_CCTV 개수.csv')
risk_df = pd.read_csv('3년치 risk_index.csv')
```

```python
plt_df
```

```python
risk_df = risk_df.sort_values(by=['지역'])
risk_df = risk_df.reset_index(drop=True)
```

```python
risk_df.columns = ['name', 'risk_index']
```

```python
data = pd.merge(risk_df, plt_df, on='name')
```

```python
data
```

```python
data.to_csv('risk_df.csv', encoding='utf-8', index=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
from matplotlib import font_manager, rc

rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(200,200))
sns.pairplot(data,
            x_vars=['시설개수', 'CCTV설치대수'],
            y_vars=['risk_index'], kind='reg', size=5)
sns.set_style({'font.family':'serif', 'font.serif':['Times New Roman']})
plt.show()
```

```python
plt.figure(figsize=(20, 10))
target_col=['시설개수', 'CCTV설치대수', 'risk_index']
sns.heatmap(data[target_col], annot=True, fmt='f', linewidths=.5)
plt.title('risk_late')
plt.show()
```
