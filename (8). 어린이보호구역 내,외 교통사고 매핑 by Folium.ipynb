```python
import folium
import pandas as pd
```

```python
df_300 = pd.read_csv('어린이보호구역 내 어린이보행사고.csv', encoding='cp949')
```

```python
df_500 = pd.read_csv('어린이보호구역 외 어린이보행사고.csv', encoding='cp949')
```

```python
center = [37.577185, 126.938749] 
m = folium.Map(location=center, zoom_start=13, tiles = 'Stamen Terrain')
    
m.save('map.html')
```

```python
# ---------- 300m · 500m 반경 어린이보행자 사고 위치 매핑
from folium.plugins import MarkerCluster
for i in df_300.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df_300['위도'][i]
    lat = df_300.loc[i, '위도']
    lng = df_300.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], icon=folium.Icon(color='blue',icon='top'))
    marker.add_to(m)
    marker_cluster = MarkerCluster().add_to(m)

m.save('보호구역 내 사고.html')
```

```python
for i in df_500.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = df_500['위도'][i]
    lat = df_500.loc[i, '위도']
    lng = df_500.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], icon=folium.Icon(color='red',icon='down'))
    marker.add_to(m)
    marker_cluster = MarkerCluster().add_to(m)

m.save('보호구역 내 사고2.html')
m
```
